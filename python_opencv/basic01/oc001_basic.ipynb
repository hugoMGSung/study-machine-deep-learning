#### 모듈 임포트

````python
import cv2
import numpy as np
````

#### 버전확인

````python
print(f'OpenCV {cv2.__version__}')
````

#### 이미지 로드
- 경로 조심!

````python
src = cv2.imread('../images/Olivia_Hussey.jpg')
````

````python
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
````

#### 화면 표시
- 이미지 사이즈가 너무 커서 전부 나타나지 않음

````python
cv2.imshow("original", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#### 흑백 및 Histogram 처리

````python
result = np.ones((src.shape[0], 256), dtype=np.uint8)
# 히스토그램 계산함수
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
cv2.normalize(hist, hist, 0, 256, cv2.NORM_MINMAX)
````

````python
for x, y in enumerate(hist):
    cv2.line(result, (x, src.shape[0]), (x, src.shape[0]), 255, 2)
````

````python
dst = np.hstack([src[:, :, 0], result])
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
